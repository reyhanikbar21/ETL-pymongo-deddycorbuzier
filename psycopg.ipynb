# Insert Data to Postgres

```python
#import necessary library
import pandas as pd
import numpy as np
import re
from datetime import datetime, timedelta
import psycopg2
from psycopg2 import sql
```

```python
# load csv file
df=pd.read_csv("DeddyVideo.csv")
df=df.rename(columns={
    "_id" : "video_id"}
)
df
```

Change the "publishedAt" date format.

```python
def change_date_format(date):
    old_format_date = datetime.fromisoformat(date)
    new_format_date = old_format_date.strftime("%Y-%m-%d %H:%M:%S")
    return new_format_date
```

```python
df1=df
df1["publishedAt"]=df1["publishedAt"].apply(change_date_format)
df1
```

Change the "duration" format

```python
def change_duration_format(duration):
    pattern = r'PT(\d+H)?(\d+M)?(\d+S)?'
    matches = re.search(pattern, duration)
    hours = matches.group(1) if matches.group(1) else 0
    minutes = matches.group(2) if matches.group(2) else 0
    seconds = matches.group(3) if matches.group(3) else 0
    hours = int(hours[:-1]) if hours else 0
    minutes = int(minutes[:-1]) if minutes else 0
    seconds = int(seconds[:-1]) if seconds else 0
    duration = timedelta(hours=hours, minutes=minutes, seconds=seconds)
    return str(duration)
```

```python
df2=df1
df2["duration"]=df2["duration"].apply(change_duration_format)
df2
```

drop "tags" column

```python
df3=df2.drop(columns="tags")
df3.to_csv("data_to_import.csv", index=False)
df3
```
